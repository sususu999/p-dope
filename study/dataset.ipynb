```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Excelファイルのパス
file_path = r'C:\python\study\test.xlsx'

# ExcelファイルをDataFrameとして読み込む
df = pd.read_excel(file_path)

# データの先頭5行を表示
df.head()
```

```python
x = df['濃度'].values.reshape(-1, 1)  # ここを修正
t = df[['ドーピング0回目', '1回目', '2回目', '3回目']]
```

```python
x
```

```python
t
```

```python
from sklearn.model_selection import train_test_split   #データを学習用とテスト用に分割
```

```python
x_train, x_test, t_train, t_test=train_test_split(x, t, test_size=0.3, random_state=0) #学習用とテスト用に分割
```

```python
len(x_test) #テスト用データの数を確認
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model= LinearRegression() 
```

```python
model.fit(x_train, t_train)
```

```python
model.coef_
```

```python
print(f'test score: {model.score(x_test, t_test)}') #テストデータのスコア
print(f'train score: {model.score(x_train, t_train)}') #学習データのスコア
```

```python
y = model.predict(x_test) #予測値を取得
```

```python
print(f'予測値: {y[4]}') # 予測値
print(f'実測値: {t_test.iloc[4].values}') # 実測値（1行目の全ての値）
```

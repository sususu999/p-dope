```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns #必要なライブラリをインポート（使えるようにする）
import itertools
```

```python
from sklearn.linear_model import LinearRegression #sklearnの線形回帰モデルをインポート
```

```python
from sklearn.model_selection import train_test_split #データを学習用とテスト用に分割するための関数をインポート
```

```python
df = pd.read_excel("test_result3.xlsx") #実験結果を読み込んで、dfという変数に格納
```

```python
df #dfに格納されたデータを表示
```

```python
df.columns = ['halfgold_ratio', 'doping_concentration', 'resistance_reduction'] #列名が日本語になっているので、英語に変更
```

```python
df #変更されているか確認
```

```python
x = df[['halfgold_ratio', 'doping_concentration']]
y = df['resistance_reduction'] #目的関数をyに入れる
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1) #学習用とテスト用にデータを分割（例: 80%学習, 20%テスト）
```

```python
len(x_train), len(x_test) #学習用とテスト用のデータ数を確認
```

```python
model = LinearRegression() #線形回帰モデルを作成
model.fit(x_train, y_train) #モデルに自分のデータを学習させる
```

```python
print(f'train scoer: {model.score(x_train, y_train)}') #学習用スコアと予測したモデルの比較
print(f'test scoer: {model.score(x_test, y_test)}') #テスト用スコアと予測したモデルの比較　♯1がMAXだけど0.37は低すぎるね。データが少ないから？
```

```python
t= model.predict(x_test) #テスト用データの予測値を表示
```

```python
print(f'予測値: {t[1]}')              # テスト用データの予測値を表示 #抵抗値がマイナスの域に行く！すごい！
print(f'実測値: {y_test.iloc[1]}')    # テスト用データの実測値を表示
```

```python
# テスト用データの予測値と実測値を一覧で表示
for i in range(len(t)):
    print(f'予測値: {t[i]:.3f} / 実測値: {y_test.iloc[i]:.3f}')
```

```python
# 半金比率: 90,80,...,0（10刻み）
halfgold_vals = list(range(90, -1, -10))

# ドーピング濃度: 0.5〜16を1刻みで生成
doping_vals = np.round(np.arange(0.5, 16.0 + 1e-9, 1.0), 3)

# 組み合わせ作成
predict_df = pd.DataFrame(list(itertools.product(halfgold_vals, doping_vals)),
                          columns=['halfgold_ratio', 'doping_concentration'])

# model が存在するか確認
if 'model' not in globals():
    raise RuntimeError("model が定義されていません。先に学習を実行してください。")

# 予測
predict_df['predicted_resistance_reduction'] = model.predict(predict_df)

# 保存
out_path = "predictions_half_90to0_doping_0.5to16.xlsx"
predict_df.to_excel(out_path, index=False)
print("保存しました:", out_path)
```
